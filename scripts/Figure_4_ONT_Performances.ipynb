```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='ticks')
```

```python
df = pd.read_excel("../data/Tool_Performance/ONT_Simulations.xlsx")
df_raw = df[df['Complexity']=="raw"]
df_assembly = df[df['Complexity']=="assemblies"]
```

### Tool Performance on Raw Reads

```python
_tools= ['VirSorter', 'VirFinder', 'Earth Virome', 'Kraken2', "DeepVirFinder", "VirBrant"]
c = ["gold", "green", "violet", "maroon", "orange", "dodgerblue"]

my_dict = dict(zip(_tools, c))


fg = sns.FacetGrid(data=df_raw, hue='Tool', hue_order=_tools, palette=my_dict, aspect=1.61)
fg.map(plt.scatter, 'Recall', 'Precision').add_legend()

plt.xlim(0, 1.05) 
plt.ylim(0, 1.05) 
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Performance of Tools on Raw NanoPore Simulation Reads")
plt.savefig("../figures/Figure_4_ONT_Simulation/Figure_4A_Performance_of_Tools_on_Raw_ONT_Reads.jpg")
```

### Tool Performance on Assembled Contigs

```python
_tools= ['VirSorter', 'VirFinder', 'Earth Virome', 'Kraken2', "DeepVirFinder", "VirBrant"]

fg = sns.FacetGrid(data=df_assembly, hue='Tool', hue_order=_tools, palette=my_dict, aspect=1.61)
fg.map(plt.scatter, 'Recall', 'Precision').add_legend()


plt.xlim(0, 1.05) 
plt.ylim(0, 1.05) 
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Performance of Tools on Assemblied NanoPore Simulation")
plt.savefig("../figures/Figure_4_ONT_Simulation/Figure_4B_Performance_of_Tools_on_Assembled_ONT.jpg")
```
